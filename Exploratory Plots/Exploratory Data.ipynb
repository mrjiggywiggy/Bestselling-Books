```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("../Preprocessing/preprocessdffixed.csv")
dataframe = dataframe.drop(['Unnamed: 0' ], axis=1)
dataframe
```

```python
plt.style.use('default')
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(7, 3.5))

ax.scatter(dataframe['Price'], dataframe['User Rating'], color='k')

ax.set_ylabel('User Rating', fontsize=14)
ax.set_xlabel('Price', fontsize=14)
ax.set_title('Price vs User Rating', fontsize=14)

fig.tight_layout()

# higher priced books get higher ratings

```

```python
plt.style.use('default')
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(7, 3.5))

ax.scatter(dataframe['Genre'], dataframe['User Rating'], color='k')
ax.set_title('Genre vs User Rating', fontsize=14)

ax.set_ylabel('User Rating', fontsize=14)
ax.set_xlabel('Genre', fontsize=14)

fig.tight_layout()

# genre 1 gets higher rating
# non-fiction 1
# fiction 0
```

```python
plt.style.use('default')
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(7, 3.5))

ax.scatter(dataframe['Year'], dataframe['User Rating'], color='k')

ax.set_ylabel('User Rating', fontsize=14)
ax.set_xlabel('Year', fontsize=14)
ax.set_title('Year vs User Rating', fontsize=14)

fig.tight_layout()

# This shows that year has nothing to do with rating

```

```python
# Here you can see that most houses are <1H Ocean and least houses are
# on island
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
m = dataframe['Genre'].value_counts()
print(m)
labelss = ["Non Fiction","Fiction"]
ocean_values = [m[0],m[1]]
perc = ocean_values / np.sum(ocean_values)
ax.pie(perc, labels=labelss, );
ax.set_title("Genre");
```
