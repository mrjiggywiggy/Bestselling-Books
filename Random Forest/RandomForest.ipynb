```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
from sklearn.model_selection import GridSearchCV
user_ratingRMSE=[]
total_starsRMSE=[]
user_ratingR2=[]
total_starsR2=[]
```

```python
dataframe = pd.read_csv("../Preprocessing/preprocessdffixed.csv")
```

```python
dataframe = dataframe.drop(['Unnamed: 0' ], axis=1)
dataframe
```

```python
dataframe.info()
X_to_split = dataframe.drop('User Rating', axis=1).values
y_to_split = dataframe['User Rating']
y_to_split
```

```python
X_train, X_test, y_train, y_test \
    = train_test_split(X_to_split, y_to_split)
```

```python
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
y_predict = rf.predict(X_train)
print('User Rating')
print('RMSE on the training set: ', np.sqrt(MSE(y_predict, y_train)))
print('R^2 score on training set: ', rf.score(X_train,y_train))


y_test_predict = rf.predict(X_test)
y_test_predict
print('RMSE on the test set: ', np.sqrt(MSE(y_test_predict, y_test)))
print('R^2 score on test set: ', rf.score(X_test,y_test))
```
